#### I think we should increase the number of imgs from "disguted" class in train data

```python
disguted_datagn = ImageDataGenerator(horizontal_flip=True,
                                   width_shift_range=0.1,
                                   height_shift_range=0.05,
                                   rescale = 1./255)
i=0
images=os.listdir("data/train/disgusted")
for img in images:
    img=cv.imread("data/train/disgusted/"+img)
    img=img.reshape((1,)+img.shape)
    for k in disguted_datagn.flow(img,batch_size = 16,save_to_dir="data/train/disgusted")  :
        i+=1
        if i > 5:
            break

    
```

### Model

```python
CNN= Sequential()
CNN.add(Conv2D(32,(3,3),padding='same',activation='relu',input_shape=(48,48,3)))
CNN.add(Conv2D(64,(3,3),padding='same',activation='relu'))
CNN.add(BatchNormalization())
CNN.add(MaxPool2D(pool_size=(2, 2)))
CNN.add(Dropout(0.25))

CNN.add(Conv2D(128,(3,3),padding='same',activation='relu'))
CNN.add(Conv2D(256,(3,3),padding='same',activation='relu'))
CNN.add(BatchNormalization())
CNN.add(MaxPool2D(pool_size=(2, 2)))
CNN.add(Dropout(0.25))

CNN.add(Flatten()) 
CNN.add(Dense(50,activation = 'tanh'))

CNN.add(Dense(25,activation = 'tanh'))

CNN.add(Dense(20,activation = 'tanh'))

CNN.add(BatchNormalization())
CNN.add(Dropout(0.25))

CNN.add(Dense(7, activation='softmax'))
```

```python
CNN.summary()
```

```python
train=ImageDataGenerator()
test=ImageDataGenerator()
train =train.flow_from_directory("data/train/",class_mode="categorical")
test=test.flow_from_directory("data/test/",class_mode="categorical")
```

```python
CNN.compile(
    optimizer = keras.optimizers.Adam(lr=0.0005), 
    loss='categorical_crossentropy', 
    metrics=['accuracy']
  )
```

```python
history = CNN.fit(
    train ,
    epochs=25,batch_size=1)
```
